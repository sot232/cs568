```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('books.csv')
```

```python
print(df.shape)
```

```python
df.head(3)
```

```python
df['rating'].value_counts()
```

```python
def rating_to_int(x):
    if x == 'One':
        return 1
    elif x == 'Two':
        return 2
    elif x == 'Three':
        return 3
    elif x == 'Four':
        return 4
    elif x == 'Five':
        return 5
df['rating'] = df['rating'].apply(lambda x: rating_to_int(x))
```

```python
df.plot.scatter(x='price', y='rating')
```

```python
df['price'].hist()
```

```python
df['stock'].value_counts()
```

```python
# adjust line size labels on the y-axis 

# Method 1: Using matplotlib with line spacing control
fig, ax = plt.subplots(figsize=(10, 8))
df.plot.scatter(x='price', y='category', ax=ax)

# Adjust line spacing between y-axis labels
ax.tick_params(axis='y', labelsize=10)  # Adjust font size
plt.setp(ax.get_yticklabels(), linespacing=1.5)  # Adjust line spacing

plt.tight_layout()
plt.show()
```
